```python
import cv2
import os
import random

path = './training_hr_images/'
img_list = os.listdir(path)

for file in img_list:
    img_path = os.path.join(path, file)
#     print(file.split(".")[0])
#     print(img_path)

    image = cv2.imread(img_path)

    # Flipped Horizontally 水平翻转
    h_flip = cv2.flip(image, 1)
    tmp = file.split(".")[0] + "_flip_h.png"
    h_flip_path = os.path.join(path, tmp)
#     print(h_flip_path)
    cv2.imwrite(h_flip_path, h_flip)

    # Flipped Vertically 垂直翻转
    v_flip = cv2.flip(image, 0)
    tmp = file.split(".")[0] + "_flip_v.png"
    v_flip_path = os.path.join(path, tmp)
#     print(v_flip_path)
    cv2.imwrite(v_flip_path, v_flip)

    # Flipped Horizontally & Vertically 水平垂直翻转
    hv_flip = cv2.flip(image, -1)
    tmp = file.split(".")[0] + "_flip_hv.png"
    hv_flip_path = os.path.join(path, tmp)
#     print(hv_flip_path)
    cv2.imwrite(hv_flip_path, hv_flip)
    
    y1 = random.randint(1,15)  # 20
    x1 = random.randint(1,15)  # 20
    y2 = random.randint(20,image.shape[0])  # 80
    x2 = random.randint(20,image.shape[1])  # 80
    crop_1 = image[y1:y2, x1:x2]
    tmp = file.split(".")[0] + "_crop_1.png"
    crop_path_1 = os.path.join(path, tmp)
#     print(crop_path_1)
    cv2.imwrite(crop_path_1, crop_1)
    
    y1 = random.randint(15,30)
    x1 = random.randint(15,30)
    y2 = random.randint(50,image.shape[0])
    x2 = random.randint(50,image.shape[1])
    crop_2 = image[y1:y2, x1:x2]
    tmp = file.split(".")[0] + "_crop_2.png"
    crop_path_2 = os.path.join(path, tmp)
#     print(crop_path_2)
    cv2.imwrite(crop_path_2, crop_2)
    
    y1 = random.randint(30,40)
    x1 = random.randint(30,40)
    y2 = random.randint(70,image.shape[0])
    x2 = random.randint(70,image.shape[1])
    crop_3 = image[y1:y2, x1:x2]
    tmp = file.split(".")[0] + "_crop_3.png"
    crop_path_3 = os.path.join(path, tmp)
#     print(crop_path_3)
    cv2.imwrite(crop_path_3, crop_3)
    
    y1 = random.randint(1,40)
    x1 = random.randint(1,40)
    y2 = random.randint(70,image.shape[0])
    x2 = random.randint(70,image.shape[1])
    crop_4 = image[y1:y2, x1:x2]
    tmp = file.split(".")[0] + "_crop_4.png"
    crop_path_4 = os.path.join(path, tmp)
#     print(crop_path_4)
    cv2.imwrite(crop_path_4, crop_4)
    
    short = image.shape[0]
    if image.shape[1] < short:
        short = image.shape[1]
    resized_1 = cv2.resize(image, (short, short))
    tmp = file.split(".")[0] + "_resize_1.png"
    resize_path_1 = os.path.join(path, tmp)
#     print(resize_path_1)
    cv2.imwrite(resize_path_1, resized_1)
    
#     long = image.shape[0]
#     if image.shape[1] > long:
#         long = image.shape[1]
#     resized_2 = cv2.resize(image, (long, long))
#     tmp = file.split(".")[0] + "_resize_2.png"
#     resize_path_2 = os.path.join(path, tmp)
# #     print(resize_path_2)
#     cv2.imwrite(resize_path_2, resized_2)
    
    # Rotation
    M=cv2.getRotationMatrix2D((image.shape[1]/2,image.shape[0]/2), 90, 1)
    img_ro_1 =cv2.warpAffine(image, M, image.shape[:2])
    tmp = file.split(".")[0] + "_rotation_1.png"
    rotation_path_1 = os.path.join(path, tmp)
#     print(resize_path_1)
    cv2.imwrite(rotation_path_1, img_ro_1)
    
    # Rotation
    M=cv2.getRotationMatrix2D((image.shape[1]/2,image.shape[0]/2), 180, 1)
    img_ro_2 =cv2.warpAffine(image, M, image.shape[:2])
    tmp = file.split(".")[0] + "_rotation_2.png"
    rotation_path_2 = os.path.join(path, tmp)
#     print(resize_path_2)
    cv2.imwrite(rotation_path_2, img_ro_2)
```
